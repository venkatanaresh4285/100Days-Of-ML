# Matplotlib

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0, 4*np.pi, 100)

# Defining Wave
sin_y = np.sin(x)
cos_y = np.cos(x)

# Plotting the Wave
plt.plot(x, sin_y, label = 'Sine')
plt.plot(x, cos_y, label = 'Cosine')

plt.legend(loc = 'upper right')
plt.title('Sine Wave')

plt.show()
```

```python
plt.plot(x, sin_y,  '*')
plt.plot(x, cos_y,  '--')

plt.title('Sine Wave')

plt.show()
```

```python
print(plt.style.available)
```

```python
plt.style.use('seaborn-dark')

plt.plot(x, sin_y,label = 'Sine')
plt.plot(x, cos_y,label = 'Cosine')

plt.title('Sine & Cosine Wave')

plt.legend()
plt.show()
```

```python
plt.figure(figsize = (18,5))

plt.style.use('grayscale')

plt.plot(x, sin_y,label = 'Sine')
plt.plot(x, cos_y,label = 'Cosine')

plt.title('Sine & Cosine Wave')

plt.legend()
plt.show()
```

# 3. Saving the Plots in our Local Machine

```python
plt.figure(figsize = (18,5))

plt.style.use('seaborn-dark')

fig = plt.figure()

plt.plot(x, sin_y,label = 'Sine')
plt.plot(x, cos_y,label = 'Cosine')

plt.title('Sine & Cosine Wave')

fig.savefig('Sine & Cosine.png')
```

# Scatter

```python
n = 500
x = np.random.randn(n)
y = np.random.randn(n)

plt.scatter(x,y, c = 'red', alpha = .5, s = 30)

plt.title('Scatter Plot')

plt.show()
```

```python
n = 1000

x1 = np.random.randn(n)
y1 = np.random.randn(n)

x2 = np.random.randn(n) + 2
y2 = np.random.randn(n) + 2

plt.scatter(x1,y1, c = 'red',  alpha = .7, s = 10, label = 'Group - 1')

plt.scatter(x2,y2, c = 'blue', alpha = .7, s = 10, label = 'Group - 2')

plt.title('Scatter Plot with Multiple Groups')
plt.legend()
plt.show()
```

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes  = rng.rand(100) * 1000

plt.scatter(x,y, c = colors, s = sizes , alpha = .7, cmap = 'viridis', edgecolors = 'black')

plt.title('Random Scatter Plot')
plt.xlabel('x values')
plt.ylabel('y values')

plt.colorbar()

plt.show()
```

# Histograms and Binnings

```python
data = np.random.randn(1000)

plt.hist(data)

plt.show()
```

```python
data = np.random.randn(10000000)

plt.hist(data, bins = 2000, edgecolor = 'blue', alpha = .8, color = 'blue')

plt.title('Random Histogram')
plt.ylabel('Frequency')
plt.xlabel('Values')

plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins = 20, edgecolor = 'Black', alpha = .8, color = 'red', histtype = 'stepfilled')

plt.title('Random Histogram')
plt.ylabel('Frequency')
plt.xlabel('Values')

plt.show()
```

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars)

plt.show()
```

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]
explode = [0, 0, 0 , 0, .1, 0] 

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars, explode = explode)

plt.show()
```
